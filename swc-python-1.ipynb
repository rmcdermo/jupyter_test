```python
def fahr_to_kelvin(temp_f):
    temp_k = ((temp_f-32.0)*(5.0/9.0)) + 273.15
    return temp_k
```

```python
def kelvin_to_celsius(temp_k):
    return temp_k - 273.15
```

```python
def fahr_to_celsius(temp_f):
    temp_k = fahr_to_kelvin(temp_f)
    temp_c = kelvin_to_celsius(temp_k)
    return temp_c
```

```python
def analyze(fname):
    
    data = numpy.loadtxt(fname=fname, delimiter=',')
    
    fig = matplotlib.pyplot.figure(figsize=(10,3))
    
    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)
    
    axes1.set_ylabel('mean')
    axes1.plot(numpy.mean(data, axis=0))
    
    axes2.set_ylabel('max')
    axes2.plot(numpy.max(data, axis=0))
    
    axes3.set_ylabel('min')
    axes3.plot(numpy.min(data, axis=0))
    
    fig.tight_layout()
    
    matplotlib.pyplot.show()
```

```python
analyze('inflammation-01.csv')
```

```python
def detect_problems(fname):
    
    data = numpy.loadtxt(fname=fname, delimiter=',')
    
    if (numpy.max(data, axis=0)[0] == 0) and (numpy.max(data, axis=0)[20] == 20):
        print('Suspicious maxima')
    elif numpy.sum(numpy.min(data, axis=0))==0:
        print('Suspicious mimima!!!')
    else:
        print('Everything looks awesome!')
```

```python
def center(data, desired=0):
    """
    return data centered about desired value
    center(data, desired)
    data: array of values to consider
    desired: value to center around [default = 0]
    """
    shifted = (data - numpy.mean(data)) + desired
    return shifted
```

```python
##
## z' = (z - min(z)) / (max(z) - min(z))

def fluc(data):
    mind = numpy.min(data)
    maxd = numpy.max(data)
    denom = maxd - mind
    if denom > 0:
        zp = (data - mind)/denom
    else:
        zp = numpy.zeros(numpy.shape(data))
    return zp
```

```python
## (1.0, 2.0, 3.0, 3.0) --> (0.0, 0.0, 1.0, 1.0)

def normalize_rectangle(rect):
    """
    Normalize a rectangle so that it is at the origin and 1.0 units long on its longest axis
    
    input
    -----
    rect : array of length 4 of form (x0, y0, x1, y1)
    """
    assert len(rect) == 4, 'Rectangles must contain 4 coordinates'
    
    x0, y0, x1, y1 = rect
    
    assert x0 < x1, 'Invalid x coordinates'
    assert y0 < y1, 'Invalid y coordinates'
    
    dx = x1 - x0
    dy = y1 - y0
    
    if dx > dy:
        scaled = dy/dx
        upper_x, upper_y = 1.0, scaled
    else:
        scaled = dx/dy
        upper_x, upper_y = scaled, 1.0
        
    assert 0 < upper_x and upper_x <= 1.0, 'Calculated upper x invalid'
    assert 0 < upper_y and upper_y <= 1.0, 'Calculated upper y invalid'
    
    return (0.0, 0.0, upper_x, upper_y)
```

```python
print(normalize_rectangle((0,0,1,1)))
```

```python
print(normalize_rectangle((0,0,1,)))
```

```python
print(normalize_rectangle((4,2,1,5)))
```

```python
print(normalize_rectangle((0.0,0.0,1.0,5.0)))
```

```python
print(normalize_rectangle((0.0,0.0,4.0,1.0)))
```
